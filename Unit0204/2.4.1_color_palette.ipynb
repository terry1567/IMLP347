```python
from random import choice
def aabc(start=0,end=9,length=4):
    ans=[]
    nums=range(start,end+1)
    while len(ans)<length:
        n=choice(nums)
        ans.append(n)
    return ans
a=aabc(300,400,5)
b=aabc(60,90,12)
c=aabc()
d=aabc(0,9,6)
print(a)
print(b)
print(c)
print(d)
```

```python
import requests
from pyquery import PyQuery as pq
def exchange(output="table"):
    url="https://rate.bot.com.tw/xrt?Lang=zh-TW"
    #print("爬:",url)
    res=requests.get(url)
    html=res.text
    html=pq(html)
    #<div class="hidden-phone print_show xrt-cur-indent">美金 (USD) </div>
    name_query="div.hidden-phone.print_show"
    name_list=html(name_query).text().split()
    a=[]
    for idx, name in enumerate(name_list):
        if idx%2==0:
           a.append(name)
        #print(idx,name)
    #print(a)    

#<td data-table="本行現金買入" class="rate-content-cash text-right print_hide">30.645</td>
    buy_query="td[data-table=本行現金買入].print_hide"
    buy_list=html(buy_query).text().split()
    #print(buy_list)
    #<td data-table="本行現金賣出" class="rate-content-cash text-right print_hide">31.315</td>
    sell_query="td[data-table=本行現金賣出].print_hide"
    sell_list=html(sell_query).text().split()
    #print(sell_list)
    table={}
    records=[]
    for name,buy,sell in zip(a,buy_list,sell_list):
        table[name]={"buy":buy,"sell":sell}
        records.append({"name":name,"buy":buy,"sell":sell})
    if output=="table":
        return table
    elif output=="records":
        return records
    elif output=="a":
        return a
    else:
        return f"不支此格式:{output}"
    

    
a=exchange()
b=exchange("records")
c=exchange("a")
print("a:",a)
print("b:",b)
print("c:",c)
```
